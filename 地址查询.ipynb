```python
import pandas as pd
import requests
import time
import csv
import json
path='C:/Users/夜间清风/Desktop/复杂网络分析/城市吸引力/爬取数据集/点表格.csv'
```

```python
def baidu(addr):
    url = "http://api.map.baidu.com/geocoding/v3/?" #百度地图API接口
    para = {
        "address": addr, #传入地址参数
        "output": "json",
        "ak": "GR2jncfPEYnNHrob6ZZnKOgRcStoBYCo" #百度地图开放平台申请ak
    }
    req = requests.get(url,para)
    req = req.json()
    #print(req)
    print('-' * 30)
    m = req["result"]["location"]
    g = [m['lng'],m['lat']]
    print(g)
    return g
```

```python
data=pd.read_csv(path,encoding='gbk')
datas=pd.DataFrame(data)
datas
```

```python
c=[]
for i in datas['id']:
    try:

        a=baidu(addr=i)
        b=[i]+a
        c.append(b)
    except:
        print(i+'出现错误')

```

```python
c=pd.DataFrame(c)
c.to_csv('点表格+.csv',encoding='gbk')
```
